# RPA code design
- 목표 정하기
  1. 개인정보 파일 1천개를 만들기
  1. 양식으로 정리된 CSV 파일을 목표하기
- 목표를 달성하는 데 필요한 작업 쪼개기
  1. 헤더 만들기
  1. 개인정보 내용물 사이사이에 콤마(",")를 넣어서 CSV에 맞는 형태로 정리하기

```python
# full code

#-*-coding:euc-kr

"""
Peter's Code
Clone Code w/ 6개월 치 업무를 하루만에 끝내는 업무 자동화
Last Modification : 2021.11.27
"""

# import
import time
import random
import os
import shutil

#
print("Process Start.")
start_time = time.time()

#
NUM_SAMPLES = 1000

#
alphabet_samples = "abcdefghizklmnopqrstuvwxyz123456890"

first_name_samples = "김이박최정강조윤장임"
middle_name_samples = "민서예지도하주윤채현지"
last_name_samples = "준윤우원호후서연아은진"

#
def random_string(length):
    result = ""

    for i in range(length):
        result += random.choice(alphabet_samples)

    return result

def random_name():
    result = ""

    result += random.choice(first_name_samples)
    result += random.choice(middle_name_samples)
    result += random.choice(last_name_samples)

    return result

# create dir
if os.path.exists("personal_info"):
    shutil.rmtree("personal_info")
os.mkdir("personal_info")

#
header = ", ".join(["name", "age", "e-mail", "division", "telephone", "sex"]) + "\n"

#
for i in range(NUM_SAMPLES):
    
    name = random_name()

    filename = "personal_info/" + str(i) + "_" + name + ".csv"

    #
    outfile = open(filename, "w")

    outfile.write(header)

    outfile.write(name + ", ")
    outfile.write(str(time.time())[-2:] + ", ")
    outfile.write(random_string(8) + "@bhban.com, ")
    outfile.write(random_string(3) + ", ")
    outfile.write("010-" + str(time.time())[-4:] + "-" + str(time.time())[-6:-2] + ", ")
    outfile.write(random.choice(["male", "female"]))

    #
    outfile.close()

#
print("Process Done.")
end_time = time.time()
print("Peter's code " + str(end_time - start_time) + " seconds.")
```
