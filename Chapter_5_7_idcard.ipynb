````python
%%writefile augmentation.py
#!usr/bin/bash python
#-*-coding:utf-8

import time
import os
from PIL import Image
import sys
import shutil

print("Process Start.")
start_time = time.time()

image_filename = sys.argv[1]

out_dir = "augmentation"
if out_dir not in os.listdir():
    os.mkdir(out_dir)
else:
    shutil.rmtree(out_dir)
    os.mkdir(out_dir)

image = Image.open(image_filename)
# note: using unpacking, instance
Xdim, Ydim = image.size

# init
COUNT = 1

# revised: f-formatting
# save copy of image
temp_new_file_name = f"{COUNT:05}.png"
COUNT += 1
image.save(out_dir + "/" + temp_new_file_name)
image.close()

FILELIST = [temp_new_file_name]



# Process 1: FLIP_LEFT_RIGHT
for i in range(len(FILELIST)):
    # import
    image = Image.open(out_dir + "/" + FILELIST[i])
    new_temp_name = f"{COUNT:05}.png"

    COUNT += 1
    
    # reverse
    image = image.transpose(Image.FLIP_LEFT_RIGHT)

    # save
    image.save(out_dir + "/" + new_temp_name)
    image.close()

    FILELIST.append(new_temp_name)

# Proecss 2 : FLIP_TOP_BOTTOM
for i in range(len(FILELIST)):
    # import
    image = Image.open(out_dir + "/" + FILELIST[i])
    new_temp_name = f"{COUNT:05}.png"

    COUNT += 1

    # reverse
    image = image.transpose(Image.FLIP_TOP_BOTTOM)

    # save
    image.save(out_dir + "/" + new_temp_name)
    image.close()

    FILELIST.append(new_temp_name)

# Process 3 : black & white
for i in range(len(FILELIST)):
    # import
    image = Image.open(out_dir + "/" + FILELIST[i])
    new_temp_name = f"{COUNT:05}.png"

    COUNT += 1

    # black & white
    image = image.convert("1")

    # save
    image.save(out_dir + "/" + new_temp_name)
    image.close()

    FILELIST.append(new_temp_name)

# Process 4 : rotation
for el in FILELIST:
    for i in range(180):
        # create 1000 pieces from the original
        if COUNT > 10000:
            break
        
        # import
        image = Image.open(out_dir + "/" + el)
        new_temp_name = f"{COUNT:05}.png"

        COUNT += 1

        # rotate & resize
        image = image.rotate(i+1)
        image = image.resize((Xdim, Ydim))

        # save
        image.save(out_dir + "/" + new_temp_name)
        image.close()

print("Process Done.")
end_time = time.time()
print("Peter's " + str(end_time - start_time) + " seconds.")
````

# RPA code design
- 목표 정하기
  1. 템플릿 불러오기
  1. 회사 로고 삽입하기
  1. 회사 홈페이지 url 기재하기
  1. 사원 사진 기재하기
  1. 사원 이름과 부서 기재하기
  1. 완성된 사원증 저장하기

- 목표를 달성하는 데 필요한 작업 쪼개기
  1. 이미지 위에 다른 이미지 삽입하기
  1. 이미지에 글자 기재하기

````python
%%writefile idcard.py
#!usr/bin/bash python
#-*-coding:utf-8

"""
Peter's Code
Clone Code w/ 6개월 치 업무를 하루만에 끝내는 업무 자동화
Last Modification : 2021.12.18
"""

import time
import os
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
import sys
import shutil

print("Process Start.")
start_time = time.time()

member_photo = sys.argv[1]
personal_IDs = sys.argv[2]
logo_filename = sys.argv[3]
template_filename = sys.argv[4]

# create white template if not exists
try:
    template = Image.open(template_filename)
except:
    template = Image.new("RGBA", (800, 1268), 'white')

Xdim, Ydim = template.size

url = "uos.ac.kr"

out_dir = "idcards"
if out_dir in os.listdir():
    shutil.rmtree(out_dir)

os.mkdir(out_dir)

# logo
logo = Image.open(logo_filename)
logo_x, logo_y = logo.size

# ID photos
photos = os.listdir(member_photo)
PHOTOS = []

for el in photos:
    if el.strip().split(".")[-1] not in "PNG JPG JPEG BMP png jpg jpeg bmp":
        continue
    PHOTOS.append(el)

# init
COUNT = 0

# edit logo size
new_logo_x = int(Xdim * 0.2)
new_logo_y = int(logo_y * (new_logo_x / logo_x))

resized_logo = logo.resize((new_logo_x, new_logo_y))

logo.close()

# read IDs
IDs = open(personal_IDs)

header = IDs.readline()

# insert logo
template.paste(resized_logo, (int(Xdim * 0.1), int(Ydim * 0.95 - new_logo_y)), resized_logo)

resized_logo.close()

# set Fonts
# use NanumBarunGothic
nameFont = ImageFont.truetype("/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf", 70)
smallFont = ImageFont.truetype("/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf", 40)
infoFont = ImageFont.truetype("/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf", 50)

# insert URL
x_offset = int(Xdim * 0.95 - smallFont.getsize(url)[0])
y_offset = int(Ydim * 0.02)

ImageDraw.Draw(template).text(xy=(x_offset, y_offset), text=url, font=smallFont, fill="black")

# create IDs
for line in IDs:
    # split
    splt = line.strip().split(", ")

    name = splt[0]
    division = splt[3]
    
    # draw ID background
    idcard = template.copy()

    # import ID photo by photo & resize
    photo_for_id = Image.open(member_photo + "/" + PHOTOS[COUNT])

    photo_for_id = photo_for_id.resize((int(Xdim/2), int(Xdim/2 * (4/3))))

    # insert ID photo
    idcard.paste(photo_for_id, (int(Xdim/4), int(Ydim/2 - Xdim/2 * (4/3)/2)))

    # name
    # more aesthetic
    # x: align center
    temp_name = ""
    for el in name:
        temp_name += el + " "
    name = temp_name[:-1]

    x_offset = int(Xdim * 0.5 - nameFont.getsize(name)[0]/2)
    y_offset = int(Ydim * 0.8 - nameFont.getsize(name)[1])
    
    ImageDraw.Draw(idcard).text(xy=(x_offset, y_offset), text=name, font=nameFont, fill="black")

    # div
    # x: align center
    x_offset = int(Xdim * 0.5 - infoFont.getsize(division)[0]/2)
    y_offset = int(Ydim * 0.85 - infoFont.getsize(division)[1])

    ImageDraw.Draw(idcard).text(xy=(x_offset, y_offset), text=division, font=infoFont, fill="black")

    # save ID
    idcard.save(out_dir + "/" + PHOTOS[COUNT])

    idcard.close()

    COUNT += 1

# close
template.close()

print("Porcess Done.")
end_time = time.time()
print("Peter's code " + str(end_time - start_time) + " seconds.")
````
