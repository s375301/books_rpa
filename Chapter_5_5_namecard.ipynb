# RPA design code
- 목표 정하기
  1. 명함 인쇄를 위한 템플릿을 만들기
  1. 템플릿에 회사 로고와 개인정보를 기재하기
  1. 저장하기

- 목표를 달성하는 데 필요한 작업 쪼개기
  1. 백지를 불러와 템플릿으로 만들기
  1. 템플릿에 회사 로고를 삽입하기
  1. 파이썬으로 이미지에 글자를 삽입하기
  1. 템플릿에 회사 홈페이지 url, 사무실 주소를 삽입하기
  1. 템플릿에 사원 인적 사항을 기재하기

#### 1. 백지를 불러와 템플릿으로 만들기
- ```Image.new()``` 함수를 활용하여 배경이 될 백지를 만든다.
```python
>>> Xdim = 1039
>>> Ydim = 697
>>> Image.new("RGBA", (Xdim, Ydim), "white")
>>> Image.new("CMYK", (Xdim, Ydim), 4-tuple)
```
- 단, CMYK 이미지는 투명도 표현이 불가능하므로 RGB와 달리 png로 저장할 수 없으며, jpg 포맷으로 저장해야한다.

#### 2. 템플릿에 회사 로고를 삽입하기
- ```resize()``` 함수를 사용한다.
- 단, 명함에서는 회사 로고가 어느정도 눈에 잘 띄어야 하므로 명함의 크기를 템플릿 높이의 40%로 조절한다.

#### 3. 파이썬으로 이미지에 글자 삽입하기
- ```PIL```라이브러리의 ```ImageFont``` 모듈을 사용한다.
```python
>>> from PIL import ImageFont
```
- 사용하고자하는 글꼴, 크기, 굵기를 정하자.
```python
>>> nameFont = ImageFont.truetype("font/gulim.ttc", 80)
>>> from PIL import ImageDraw
>>> ImageDraw.Draw(My_img).text(xy=(x_offset, y_offset), text="contents", font=nameFont, fill="black")
```

````python
%%writefile namecard.py
#!usr/bin/bash python
#-*-coding:utf-8

"""
Peter's Code
Clone Code w/ 6개월 치 업무를 하루만에 끝내는 업무 자동화
Last Modification : 2021.12.16
"""

import time
import os
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
import sys
import shutil

print("Process Start.")
start_time = time.time()

personal_IDs = sys.argv[1]
logo_filename = sys.argv[2]

location = "University of Seoul 163 Seoulsiripdaero, Seoul"
url = "uos.ac.kr"

out_dir = "namecards"
if out_dir not in os.listdir():
    os.mkdir(out_dir)
else:
    shutil.rmtree(out_dir)
    os.mkdir(out_dir)

logo = Image.open(logo_filename)
# unpack
logo_x, logo_y = logo.size

Xdim = 1039
Ydim = 697

# resize logo
new_logo_y = int(Ydim * 0.4)
new_logo_x = int(logo_x * (new_logo_y / logo_y))

resized_logo = logo.resize((new_logo_x, new_logo_y))
logo.close()

IDs = open(personal_IDs)

header = IDs.readline()

# png: RGBA
image = Image.new("RGBA", (Xdim, Ydim), "white")
image.paste(resized_logo, (int(Xdim * 0.1), int(Ydim * 0.1)), resized_logo) #

resized_logo.close()

# name
nameFont = ImageFont.truetype('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', 70)
# URL & Location
smallFont = ImageFont.truetype('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', 40)
# extra infos
infoFont = ImageFont.truetype('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', 50)

# insert URL
x_offset = int(Xdim * 0.95 - smallFont.getsize(url)[0])
y_offset = int(Ydim * 0.05)
ImageDraw.Draw(image).text(xy=(x_offset, y_offset), text=url, font=smallFont, fill="black")

# insert Location
x_offset = int(Xdim * 0.95 - smallFont.getsize(location)[0])
y_offset = int(Ydim * 0.95 - smallFont.getsize(location)[1])
ImageDraw.Draw(image).text(xy=(x_offset, y_offset), text=location, font=smallFont, fill="black")

# insert infos
for line in IDs:

    # split CSV file by Commas
    splt = line.strip().split(", ")

    name = splt[0]
    e_mail = splt[2]
    division = splt[3]
    telephone = splt[4]

    namecard = image.copy()

    # more aesthetic
    temp_name = ""
    for el in name:
        temp_name += el + " "
    name = temp_name[:-1]

    # insert name
    x_offset = int(Xdim * 0.9 - nameFont.getsize(name)[0])
    y_offset = int(Ydim * 0.4 - nameFont.getsize(name)[1])

    ImageDraw.Draw(namecard).text(xy=(x_offset, y_offset), text=name, font=nameFont, fill="black")

    # insert division
    x_offset = int(Xdim * 0.9 - infoFont.getsize(division)[0])
    y_offset = int(Ydim * 0.5 - infoFont.getsize(division)[1])

    ImageDraw.Draw(namecard).text(xy=(x_offset, y_offset), text=division, font=infoFont, fill="black")

    # insert telephone number
    x_offset = int(Xdim * 0.9 - infoFont.getsize(telephone)[0])
    y_offset = int(Ydim * 0.65 - infoFont.getsize(telephone)[1])

    ImageDraw.Draw(namecard).text(xy=(x_offset, y_offset), text=telephone, font=infoFont, fill="black")

    # insert e-mail
    x_offset = int(Xdim * 0.9 - infoFont.getsize(e_mail)[0])
    y_offset = int(Ydim * 0.75 - infoFont.getsize(e_mail)[1])

    ImageDraw.Draw(namecard).text(xy=(x_offset, y_offset), text=e_mail, font=infoFont, fill="black")


    namecard.save(out_dir + "/" + division + "_" + name + "_" + telephone + ".png")

    namecard.close()

image.close()

print("Process Done.")
end_time = time.time()
print("Peter's code " + str(end_time - start_time) + " seconds.")
````
