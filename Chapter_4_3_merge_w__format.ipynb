````python
# create CSVs

#-*-coding:euc-kr

"""
Peter's Code
Clone Code w/ 6개월 치 업무를 하루만에 끝내는 업무 자동화
Last Modification : 2021.11.27
"""

# import
import time
import random
import os
import shutil

#
print("Process Start.")
start_time = time.time()

#
NUM_SAMPLES = 1000

#
alphabet_samples = "abcdefghizklmnopqrstuvwxyz123456890"

first_name_samples = "김이박최정강조윤장임"
middle_name_samples = "민서예지도하주윤채현지"
last_name_samples = "준윤우원호후서연아은진"

#
def random_string(length):
    result = ""

    for i in range(length):
        result += random.choice(alphabet_samples)

    return result

def random_name():
    result = ""

    result += random.choice(first_name_samples)
    result += random.choice(middle_name_samples)
    result += random.choice(last_name_samples)

    return result

# create dir
if os.path.exists("personal_info"):
    shutil.rmtree("personal_info")
os.mkdir("personal_info")

#
header = ", ".join(["name", "age", "e-mail", "division", "telephone", "sex"]) + "\n"

#
for i in range(NUM_SAMPLES):
    
    name = random_name()

    filename = "personal_info/" + str(i) + "_" + name + ".csv"

    #
    outfile = open(filename, "w")

    outfile.write(header)

    outfile.write(name + ", ")
    outfile.write(str(time.time())[-2:] + ", ")
    outfile.write(random_string(8) + "@bhban.com, ")
    outfile.write(random_string(3) + ", ")
    outfile.write("010-" + str(time.time())[-4:] + "-" + str(time.time())[-6:-2] + ", ")
    outfile.write(random.choice(["male", "female"]))

    #
    outfile.close()

#
print("Process Done.")
end_time = time.time()
print("Peter's code " + str(end_time - start_time) + " seconds.")
````

# RPA code design
- 목표 정하기
  1. 파일들을 하나로 합치기
  1. 합쳐지는 과정에서 양식을 유지하기
- 목표를 달성하는 데 필요한 작업 쪼개기
  1. 헤더와 데이터를 분리하기
  1. 헤더는 파일에 한 번만 기록하고, 데이터는 하나씩 모두 기록하기

````python
# full code

#-*-coding:euc-kr

"""
Peter's Code
Clone Code w/ 6개월 치 업무를 하루만에 끝내는 업무 자동화
Last Modification : 2021.11.27
"""

# import
import time
import sys
import os

#
print("Process Start.")
start_time = time.time()

"""
# original
directory = sys.argv[1]
"""

# revised
directory = "personal_info"

#
outfile_name = "merged_ID.csv"

#
out_file = open(outfile_name, "w")

input_files = os.listdir(directory)

outfile_has_header = False

#
for filename in input_files:
    
    # exception
    if ".csv" not in filename:
        continue
    
    #
    file = open(directory + "/" + filename)

    header = file.readline()

    if not outfile_has_header:
        out_file.write(header.strip())
        outfile_has_header = True
    
    out_file.write("\n" + file.read())

    file.close()

#
out_file.close()

#
print("Process Done.")
end_time = time.time()
print("Peter's code " + str(end_time - start_time) + " seconds.")
````

## Summary
- 헤더는 가장 첫줄에 있다는 점을 이용한다.
```py
header = file.readline()
```
- RPA 코드 설계 시, 범용성이 높은 코드를 작성하는 것이 유리하다.

#### 1. 헤더 분리
- ```file.readline()``` 은 파일을 한 줄씩 읽어온다.

#### 2. 헤더를 단 한번만 입력하는 방법
- ```outfile_has_header``` 와 같은 불리언 타입 변수를 통해 단 한번만 실행시킨다.

#### 3. 데이터를 파일에 기록하는 방법
- ```file.readline()``` vs. ```file.read()```
  1. ```file.readline()```: 파일을 한줄씩 읽어온다.
  1. ```file.read()```: 파일 전체를 하나의 문자열로 읽어온다. (범용성↑)
