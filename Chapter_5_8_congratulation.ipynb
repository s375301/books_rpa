#### 2. 수료증 제작 순서에 따라 수여번호를 1씩 증가시키기
- ```for``` 문을 사용해 반복 작업을 수행한다.

````python
%%writefile congratulation.py
#!usr/bin/bash python
#-*-coding:utf-8

"""
Peter's Code
Clone Code w/ 6개월 치 업무를 하루만에 끝내는 업무 자동화
Last Modification : 2021.12.18
"""

import time
import os
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
import sys
import datetime

print("Process Start.")
start_time = time.time()

# commandline inputs
personal_IDs = sys.argv[1]
template_filename = sys.argv[2]

# open
template = Image.open(template_filename)
Xdim, Ydim = template.size

# mkdir
out_dir = "Certificates"
if out_dir not in os.listdir():
    os.mkdir(out_dir)

# IDs
IDs = open(personal_IDs)
header = IDs.readline()

# set Fonts
nameFont = ImageFont.truetype('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', 30)
dateFont = ImageFont.truetype('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', 25)
smallFont = ImageFont.truetype('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', 18)

# date
date = str(datetime.datetime.today().date())
date = date.split("-")
DATE = date[0] + "년 " + date[1] + "월 " + date[2] + "일 "

# insert date
# x: center-align
x_offset = int(Xdim / 2 - dateFont.getsize(DATE)[0]/2)
y_offset = int(Ydim * 0.7)

ImageDraw.Draw(template).text(xy=(x_offset, y_offset), text=DATE, font=dateFont, fill="black")

# init
COUNT = 0

# make individual Certificates
for line in IDs:
    # split
    splt = line.strip().split(", ")

    name = splt[0]
    division = splt[3]

    # copy template
    suryojung = template.copy()

    # name
    # more aesthetic
    temp_name = ""
    for el in name:
        temp_name += el + " "
    name = "성      명 : " + temp_name[:-1]

    # div
    # more aesthetic
    temp_division = ""
    for el in division:
        temp_division += el + " "
    division = "소속부서 : " + temp_division

    # insert name
    x_offset = int(Xdim * 0.15)
    y_offset = int(Ydim * 0.35)

    ImageDraw.Draw(suryojung).text(xy=(x_offset, y_offset), text=name, font=nameFont, fill="black")

    # insert div
    y_offset += nameFont.getsize(name)[1]*1.5

    ImageDraw.Draw(suryojung).text(xy=(x_offset, y_offset), text=division, font=nameFont, fill="black")

    # certificate number
    suyeo = f"수여번호 : {int(DATE[:4])}-{COUNT:06}"

    # insert CN
    y_offset = int(Ydim * 0.12)

    ImageDraw.Draw(suryojung).text(xy=(x_offset, y_offset), text=suyeo, font=smallFont, fill="black")

    # save
    suryojung.save(out_dir + "/" + str(COUNT) + ".png")

    suryojung.close()

    COUNT += 1

template.close()

print("Process Done.")
end_time = time.time()
print("Peter's code " + str(end_time - start_time) + " seconds.")
````
